# Bikeability usage (creation of a h3 grid)

> This is an example on how to generate h3 cells for a specific region

### Step 1: import libraries and packages

```python
import os
import sys
project_path = os.path.abspath('../')
sys.path.append(project_path)
from bikeability import grid
import matplotlib.pyplot as plt
import osmnx as ox
```

### Step 2: download the boundaries of your region of interest. In this example it is the city of Munich

```python
query = ["m√ºnchen", 1]

boundary_munich = ox.geocoder.geocode_to_gdf(query[0], which_result=query[1])
boundary_munich = boundary_munich.explode(index_parts=True).iloc[:1]
boundary_munich.plot()
plt.show()
```

### Step 3.1: generation of a h3 grid in resolution 7 (__[H3 resolutions](https://h3geo.org/docs/core-library/restable/)__)

```python
id_column = 'h3_id'

h3_test = grid.create_h3_grid(boundary_munich, res=7)
h3_test.plot()
h3_test = h3_test.reset_index(names=id_column)
plt.show()
```

### Step 3.2: generation of a h3 grid in resolution 8

```python
h3_test = grid.create_h3_grid(boundary_munich, res=8)
h3_test.plot()
h3_test = h3_test.reset_index(names=id_column)
plt.show()
```

### Step 3.3: generation of a h3 grid in resolution 9

```python
h3_test = grid.create_h3_grid(boundary_munich, res=9)
h3_test.plot()
h3_test = h3_test.reset_index(names=id_column)
plt.show()
```
